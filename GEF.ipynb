```python
import numpy as np
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel, pipeline
from keybert import KeyBERT


tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = TFBertModel.from_pretrained('bert-base-uncased')
kw_model = KeyBERT()


keyword_extractor = pipeline("feature-extraction", model="distilbert-base-uncased")

# Get embeddings from keywords
def get_embeddings(keywords):
    inputs = tokenizer(keywords, return_tensors='tf', padding=True, truncation=True)
    outputs = bert_model(inputs)
    return tf.reduce_mean(outputs.last_hidden_state, axis=1)

#Extract keywords from a sentence
def extract_keywords(sentence):
    keywords = kw_model.extract_keywords(sentence, top_n=3)
    return [keyword for keyword, _ in keywords]


X_train = []
y_train = []

for filter_data in filters:
    # Embeddings for keywords
    keywords = ' '.join(filter_data['keywords'])
    embedding = get_embeddings(keywords).numpy()
    X_train.append(embedding)


    parameters = [
        filter_data['brightness'],
        filter_data['exposure'],
        filter_data['vibrance'],
        filter_data['highlights'],
        filter_data['shadows'],
        filter_data['contrast'],
        filter_data['saturation'],
        filter_data['natural_saturation'],
        filter_data['color_temperature'],
        filter_data['tint'],
        filter_data['black_point'],
        filter_data['vignette'],
        filter_data['clarity']
    ]
    y_train.append(parameters)

X_train = np.array(X_train).squeeze()
y_train = np.array(y_train)

class ParameterGenerator(tf.keras.Model):
    def __init__(self, output_dim):
        super(ParameterGenerator, self).__init__()
        self.dense1 = tf.keras.layers.Dense(128, activation='relu')
        self.dense2 = tf.keras.layers.Dense(64, activation='relu')
        self.output_layer = tf.keras.layers.Dense(output_dim)

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.output_layer(x)

# Init
output_dim = 13
model = ParameterGenerator(output_dim)
model.compile(optimizer='adam', loss='mse')


model.fit(X_train, y_train, epochs=150, batch_size=8)

def generate_parameters_from_sentence(sentence):
    keywords = extract_keywords(sentence)
    print("Extracted Keywords:", keywords)
    embeddings = get_embeddings(' '.join(keywords))
    print("Input Embeddings Shape:", embeddings.shape)
    parameters = model(embeddings)
```

```python
import numpy as np
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel, pipeline
from keybert import KeyBERT

kw_model = KeyBERT()


def extract_keywords(sentence):
    keywords = kw_model.extract_keywords(sentence, top_n=3)
    return [keyword for keyword, _ in keywords]

def generate_parameters_from_sentence(sentence):
    keywords = extract_keywords(sentence)  # Extract keywords from the sentence
    print("Extracted Keywords:", keywords)
    embeddings = get_embeddings(' '.join(keywords))
    print("Input Embeddings Shape:", embeddings.shape)
    parameters = loaded_model(embeddings)  # Use the loaded model
    return parameters.numpy()



def interpret_instruction(instruction):
    instruction = instruction.lower()
    for key in action_map:
        if key in instruction:
            return action_map[key]["parameter"], action_map[key]["change"]

    return None, 0

def update_parameters(current_params, instruction):
    param, change = interpret_instruction(instruction)

    if param:
        current_params[filterParam[param]] += change
    else:
        generated_params = generate_parameters_from_sentence(instruction)
        current_params = np.clip(generated_params[0], -10, 10)  # Limit range

    return current_params

sentence = "Neon light with a vintage tint"
generated_params = generate_parameters_from_sentence(sentence)
generated_params = generated_params.astype(int)
updated_params= generated_params.tolist()
print("Generated Parameters:", updated_params)

current_params = np.array(updated_params[0])

instruction = "add a dreamy effect."
print("Generated Parameters:", current_params)

# Update parameters
new_params = update_parameters(current_params, instruction)
new_params = new_params.astype(int)
updated_params =new_params.tolist()
print("Updated Parameters:", updated_params)
```
